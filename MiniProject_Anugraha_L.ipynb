```
import os
import numpy as np
datadir = '/content/drive/MyDrive/Mini_Project'

categories = ['Peacocks','Tigers'] 
target = []
flat_data = []

for i in categories:
  path = os.path.join(datadir,i)             # Creating a path for classes
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    img_resize = resize(img_array,(150,150,3))    # Normalizing the values
    flat_data.append(np.ndarray.flatten(img_resize))
    target.append(i)

flat_data = np.array(flat_data)
target = np.array(target)
```

```
df = pd.DataFrame(flat_data)
df['Target'] = target
df
```

```
x = df.iloc[:,0:67500]
y = df.iloc[:,-1]
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state=0,stratify=y)
```

```
print(y_train.shape)
print(y_test.shape)
```

```
np.unique(y_train,return_counts=True)
```

```
np.unique(y_test,return_counts=True)
```

```
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param = [
    {'C':[1,10,100,1000],'kernel':['linear','rbf','poly']},
    {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['linear','rbf','poly']}
]

svc = SVC(probability = True)
grid = GridSearchCV(svc,param)
grid.fit(x_train,y_train)
```

```
y_pred = grid.predict(x_test)
y_pred
```

```
y_test
```

```
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_pred,y_test)
```

```
confusion_matrix(y_pred,y_test)
```

```
print(classification_report(y_pred,y_test))
```

```
# Saving the model
import pickle 
pickle.dump(grid,open('img_model.p','wb'))
```
